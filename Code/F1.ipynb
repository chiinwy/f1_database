# Recorrer los nombres y descargar la primera imagen de cada búsqueda
for c in cons:
    busqueda = f"https://www.google.com/search?q={c} f1 logo"
    response = requests.get(busqueda)
    response.raise_for_status()

    soup = BeautifulSoup(response.text, 'html.parser')
    imagen_tag = soup.find('img')
    imagen_url = imagen_tag['src']

    # Obtener el nombre de la imagen
    nombre_imagen = c.replace(' ', '_') + '.jpg'
    ruta_imagen = os.path.join(carpeta, nombre_imagen)

    # Descargar y guardar la imagen
    imagen_response = requests.get(imagen_url)
    with open(ruta_imagen, 'wb') as f:
        f.write(imagen_response.content)
